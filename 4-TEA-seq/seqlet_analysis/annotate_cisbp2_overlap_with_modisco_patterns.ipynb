```python
cisbp_anno = pd.read_table(os.path.join(\
    path_cisbp2,
    "TF_Information_all_motifs.txt"))

cisbp_anno
```

```python
map_cisbp_mo = cisbp_anno.loc[cisbp_anno["Motif_ID"] != "."]
unique_ids = map_cisbp_mo["Motif_ID"].unique()
map_cisbp_mo = pd.Series(\
    ["|".join(map_cisbp_mo.loc[\
        map_cisbp_mo["Motif_ID"] == i,
        "TF_Name"].values) for i in unique_ids], 
    index=unique_ids)
map_cisbp_mo["None"] = "None"
map_cisbp_mo
```

```python
path_html_files = "output/chrombpnet/modisco_merged_results/fold_0/"\
    "modisco_cisbp2_html_reports"

list_modisco_html_files = os.listdir(path_html_files)
list_modisco_html_files = [i for i in list_modisco_html_files if not \
    i.startswith(".")]
```

```python
def get_modisco_ref_table_results(path_modisco_file):
    with open(path_modisco_file) as fp:
        soup = BeautifulSoup(fp, "html.parser")
    
    # Get the header for the table
    tmp_header = [i for i in soup.table.thead.text.split("\n") if i != '']

    # Get the table body 
    tmp_table = soup.findAll('tbody')[0]

    table_result = []
    for tmp_row in tmp_table.findAll('tr'):
        tmp_row_result = []
        for tmp_item in tmp_row.findAll('td'):
            # If the table value here has the image, save the src
            if len(tmp_item.findAll('img')) > 0:
                tmp_row_result.append(tmp_item.img['src'])
            # Otherwise save the text of the HTML element
            else:
                tmp_row_result.append(tmp_item.text)

        table_result.append(tmp_row_result)

    table_result = pd.DataFrame(table_result, columns=tmp_header)
    return(table_result)
```

```python
modicts = {}

for tmp_item in list_modisco_html_files:
    cname = tmp_item.split("_cisbp2_")[0]
    modicts[cname] = get_modisco_ref_table_results(os.path.join(\
        path_html_files, tmp_item))
```

```python
combined = []
cols_of_interest = ['pattern', 'num_seqlets', 'match0', 'qval0', 'match1', 
    'qval1', 'match2', 'qval2']
for tmp_cluster in modicts:
    tmp_df = modicts[tmp_cluster][cols_of_interest].copy()
    tmp_df["cluster"] = tmp_cluster
    combined.append(tmp_df)

combined = pd.concat(combined)
combined["pattern_type"] = [i[:3] for i in combined["pattern"].values]
combined["pattern_num"] = [i.split(".")[-1] for i in combined["pattern"].values]
combined["pattern_id"] = combined["cluster"] + "__" + combined["pattern_num"]
pos_anno = combined.loc[combined["pattern_type"] == "pos", ["pattern_id", 
    "num_seqlets", "match0", "qval0", "match1", "qval1", "match2", "qval2"]]
pos_anno.index = list(range(pos_anno.shape[0]))
```

```python
pos_anno["TF"] = [map_cisbp_mo[i] for i in pos_anno["match1"].values]
pos_anno
```

```python
pos_anno.to_feather("output/chrombpnet/modisco_merged_results/fold_0/"\
    "scan_modisco_pwms/annotate_modisco_pos_patterns_with_cisbp2.fea")
```
