```python
import os
import pandas as pd
import numpy as np

os.chdir("/data/salomonis2/LabFiles/Kyle/Analysis/2023_06_12_tea_seq_atac_processing/")
```

```python
# Read in the R7 annotations for the TEA-seq data
path_tea_anno = "/data/salomonis2/LabFiles/Kyle/Analysis/"\
    "2023_06_11_redo_harmony_knn_label_transfer_to_tea_seq/"\
    "output/tea_seq_r7_label_transfer_from_cite_seq_06_2023.csv"

tea_anno = pd.read_csv(path_tea_anno)
tea_anno[["barcode", "port"]] = [i.split(".") for i in tea_anno["cell"].values]
tea_anno
```

```python
# Split the annotation file by port and save as tsv file
for tmp_port in tea_anno["port"].unique():
    seg_anno = tea_anno.loc[tea_anno["port"] == tmp_port]
    seg_anno[["barcode", "R7"]].to_csv(\
        "output/tea_r7_anno_by_port/tea_seq_{}_r7_annotation.tsv".format(tmp_port),
        sep="\t", header=True, index=False)
```
